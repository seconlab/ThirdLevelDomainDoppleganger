```python
#create huge file....
###### import sys
import multiprocessing
import pyodbc
import subprocess
import io
import time
from itertools import combinations

cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=localhost;DATABASE=DomainResearch;UID=dr;PWD=dr2020')
cnxn2 = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=localhost;DATABASE=DomainResearch;UID=dr;PWD=dr2020')
#sql = "select * from orbisData"
cursor = cnxn.cursor()
cursor2 = cnxn2.cursor()
#cursor.execute(sql)

outFile = open('/mnt/e/bigChongus', 'w+t')

#num_workers = multiprocessing.cpu_count() 
#print("workers: ", str(num_workers))

tic = time.perf_counter()

secondlevelnameselect = "select secondlevelname from secondlevelcounterfaster order by count desc"
cursor.execute(secondlevelnameselect)

count = 0

while True:
    row = cursor.fetchone()
    if not row:
        break
    # Create a multiprocessing pool and iterate over domains
    # initializing string  
    test_str = row[0]
  
    # printing original string  
    #print("The original string is : " + str(test_str)) 
  
    # Get all substrings of string 
    # Using itertools.combinations() 
    res = [test_str[x:y] for x, y in combinations( 
            range(len(test_str) + 1), r = 2)]
    res.sort()
    for x in res:
        if (len(x) > 1):
            outFile.write(x + "," + row[0] + "\n")
            #outfile.write("\n")
            #print(x,row[0],sep=",")
    # printing result  
    #print("All substrings of string are : " + str(res))
    #break
    count =  count + 1
    if (count % 1000000 == 0):
        print(".")

outFile.close()
cnxn.close()
cnxn2.close()

toc = time.perf_counter()

print(f"domain names grepped in {toc - tic:0.4f} seconds")
```
