# This is a C# function to do a lockstep comparison between two large files.

* they must be sorted using unix sort
* I'll repeat it again, because it is very important.....
* they must be sorted using unix sort
* it mentions domains in the code, but this version just does a comparison to each line, and spits out the matches.

```
using System;
using System.IO;
```

```

FileStream fs = File.OpenRead("/mnt/d/dash_sorted.csv");
FileStream fsCambridge = File.OpenRead("/mnt/d/flat_domain_all_unique_sorted");

FileStream fsOutput = File.OpenWrite("/mnt/d/dash_matched_deleteme.csv");

StreamReader leftFile = new StreamReader(fs);
StreamReader rightFile = new StreamReader(fsCambridge);

StreamWriter outFile = new StreamWriter(fsOutput);

if (leftFile.EndOfStream || rightFile.EndOfStream)
{
    System.Console.WriteLine("ERROR OPENING FILE");
    return;
}


string leftDomain = null;  //delcare the variable once.
string rightDomain = null;

int compareValue = Int32.MinValue;

System.Globalization.CompareOptions myOptions = System.Globalization.CompareOptions.StringSort;
System.Globalization.CompareInfo ci = System.Globalization.CompareInfo.GetCompareInfo("en-US");


while (!leftFile.EndOfStream)
{
    //if we reach the end of the rightFile first, we have a problem....that shouldn't happen
    if (rightFile.EndOfStream)
    {
        System.Console.WriteLine("reached end of right file.  This should not happen");
        return;
    }

    //left file is 1 domain per line, nothing else.
    if (leftDomain == null)
    {
        leftDomain = leftFile.ReadLine().ToLower();
    }

    //right is in the format of 

    if (rightDomain == null)
    {
        rightDomain = rightFile.ReadLine().ToLower();
    }

    compareValue = ci.Compare(leftDomain, rightDomain, myOptions);

    if (compareValue == 0)
    {
        try
        {
            //write it to the console
            outFile.WriteLine(leftDomain);
            
            //write the same thing to the output file
            System.Console.WriteLine(leftDomain);
        }
        catch (Exception exc)
        {
            System.Console.WriteLine(exc);
        }

        //clear the right, since there may be more instances on the right
        rightDomain = null;
    }
    else if (compareValue < 0)
    {
        //go to the next value of right
        leftDomain = null;
    }
    else //if (compareValue > 0)
    {
        rightDomain = null;
    }
}

outFile.Close();
leftFile.Close();
rightFile.Close();
```
